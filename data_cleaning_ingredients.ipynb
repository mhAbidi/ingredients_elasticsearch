```python
import numpy as np
ingredients_string = np.char.replace(ingredients_string,r"\([^()]*\)","")
```

```python
ing_str=""
for i in ingredients_string:
  for j in i:
    ing_str += str(j).replace("\n",",")
```

```python
 ing_str1 = re.sub("\(.*?\)","()",ing_str) 
 
```

```python
 ing_str1 = re.sub("\[.*?\]","()",ing_str1) 
```

```python
 ing_str1 = re.sub("\{.*?\}","()",ing_str1) 
```

```python
ing_str1 = ing_str1.replace("(","")
ing_str1 = ing_str1.replace(")","")
ing_str1 = ing_str1.replace("[","")
ing_str1 = ing_str1.replace("]","")
ing_str1 = ing_str1.replace("{","")
ing_str1 = ing_str1.replace("}","")
```

```python
res = ing_str1.split(',')
print(res)
```

```python
list_set = set(res)
```

```python
len(list_set)
print(list_set)
```

```python
unique_ingredients = (list(list_set))
print(unique_ingredients)
```

```python
import csv
with open('/content/sample_data/unique_ingredients.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for i in unique_ingredients:
        writer.writerows([[i]])
```
